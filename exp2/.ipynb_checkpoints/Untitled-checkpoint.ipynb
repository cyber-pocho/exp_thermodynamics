```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import t as t_dist
import warnings
```

```python
MATERIALES = {
    "Aluminio": {"archivo": "al.csv", "masa_g": 19.0},
    "Laton"   : {"archivo": "la.csv", "masa_g": 52.3},
    "Hierro"  : {"archivo": "fe.csv", "masa_g": 33.5},
}
 
CALIBRACION = {
    "archivo": "la.csv",
    "masa_g" : 80.0,
    "Cp_ref" : 0.380,    # J/(g·K) — latón
    "nombre" : "Laton",
}
 
CP_REFERENCIA = {
    "Aluminio": 0.897,
    "Laton"   : 0.380,
    "Hierro"  : 0.449,
}
 
CONFIANZA = 0.95
```

```python
def modelo_newton(t, T_amb, T0, k): 
    
    return T_amb+(T0-T_amb)*np.exp(-k*t)
def cargar_csv(archivo):
    df=pd.read_csv(archivo)
    df.columns=df.columns.str.strip()
    df=df.dropna()
    return df["tiempo_ms"].values/1000.0, df["T_celsius"].values
```

```python
def ajustar_exponencial(tiempo_s, temp):
    T_fin  = temp[-1]
    T_ini  = temp[0]
    p0     = [T_fin, T_ini, 0.01]
    bounds = ([T_fin - 15, T_ini - 15, 1e-6],
              [T_fin + 15, T_ini + 15, 10.0])
 
    popt, pcov = curve_fit(
        modelo_newton, tiempo_s, temp,
        p0=p0, bounds=bounds, maxfev=20000
    )
    perr = np.sqrt(np.diag(pcov))
    dof  = len(tiempo_s) - 3
    t_c  = t_dist.ppf(1 - (1 - CONFIANZA) / 2, dof)
    ci   = t_c * perr
    return popt, perr, ci, dof
```

```python
def estadisticos(tiempo_s, temp, popt):
    T_pred = modelo_newton(tiempo_s, *popt)
    RSS    = np.sum((temp - T_pred)**2)
    TSS    = np.sum((temp - temp.mean())**2)
    return 1 - RSS / TSS, np.sqrt(RSS / len(temp)), RSS / (len(temp) - 3)
```

```python
def calibrar(cfg):
    print("=" * 65)
    print("  FASE 1 — CALIBRACIÓN DEL CALORÍMETRO")
    print("=" * 65)
    print(f"  Referencia : {cfg['nombre']}  |  masa: {cfg['masa_g']} g  |  Cp_ref: {cfg['Cp_ref']} J/(g·K)")
 
    tiempo_s, temp = cargar_csv(cfg["archivo"])
    popt, perr, ci, dof = ajustar_exponencial(tiempo_s, temp)
    R2, RMSE, chi2_r    = estadisticos(tiempo_s, temp, popt)
 
    k_ref     = popt[2]
    k_ref_err = ci[2]
 
    m   = cfg["masa_g"]
    Cp  = cfg["Cp_ref"]
 
    # Con C_cal = 0: hA = k * m * Cp  (límite inferior de hA)
    C_cal     = 0.0
    C_cal_err = 0.0
    hA        = k_ref * (m * Cp + C_cal)
    hA_err    = k_ref_err * (m * Cp + C_cal)   # dHA/dk * dk
 
    print(f"\n  k_ref = {k_ref:.6f} ± {k_ref_err:.6f} s⁻¹  (IC {CONFIANZA*100:.0f}%)")
    print(f"  R²    = {R2:.6f}  |  RMSE = {RMSE:.4f} °C  |  χ²ᵣ = {chi2_r:.4f}")
    print(f"\n  hA    = {hA:.6f} ± {hA_err:.6f} J/(s·K)  [con C_cal = 0]")
    print(f"\n  NOTA: C_cal fijado en 0 (un solo experimento de referencia).")
    print(f"  Para separar hA y C_cal exactamente, ver bloque 2x2 al final.")
 
    return hA, hA_err, C_cal, C_cal_err
```

```python
def analizar_material(nombre, cfg_mat, hA, hA_err, C_cal, C_cal_err):
    tiempo_s, temp      = cargar_csv(cfg_mat["archivo"])
    m                   = cfg_mat["masa_g"]
    popt, _, ci, dof    = ajustar_exponencial(tiempo_s, temp)
    R2, RMSE, chi2_r    = estadisticos(tiempo_s, temp, popt)
 
    T_amb, T0, k = popt
    k_err        = ci[2]
 
    # Cp y propagación de error completa
    Cp = (hA / k - C_cal) / m
 
    dCp_dhA   =  1.0 / (m * k)
    dCp_dk    = -hA  / (m * k**2)
    dCp_dCcal = -1.0 / m
 
    Cp_err = np.sqrt(
        (dCp_dhA   * hA_err  )**2 +
        (dCp_dk    * k_err   )**2 +
        (dCp_dCcal * C_cal_err)**2
    )
 
    return {
        "nombre" : nombre,
        "masa_g" : m,
        "T_amb"  : round(T_amb,    3),
        "T0"     : round(T0,       3),
        "k"      : round(k,        6),
        "k_err"  : round(k_err,    6),
        "R2"     : round(R2,       6),
        "RMSE"   : round(RMSE,     4),
        "chi2_r" : round(chi2_r,   4),
        "dof"    : dof,
        "Cp"     : round(Cp,       4),
        "Cp_err" : round(Cp_err,   4),
    }
```

```python
if __name__ == "__main__":
 
    hA, hA_err, C_cal, C_cal_err = calibrar(CALIBRACION)
 
    print(f"\n{'='*65}")
    print("  FASE 2 — ANÁLISIS DE MATERIALES")
    print(f"{'='*65}")
 
    resultados = []
 
    for nombre, cfg in MATERIALES.items():
        res = analizar_material(nombre, cfg, hA, hA_err, C_cal, C_cal_err)
        resultados.append(res)
 
        Cp_ref_val   = CP_REFERENCIA.get(nombre)
        discrepancia = abs(res["Cp"] - Cp_ref_val) / Cp_ref_val * 100 if Cp_ref_val else float("nan")
        sigma_n      = abs(res["Cp"] - Cp_ref_val) / res["Cp_err"] if (Cp_ref_val and res["Cp_err"] > 0) else float("nan")
 
        print(f"\n  {'─'*60}")
        print(f"  {nombre}  |  masa: {res['masa_g']} g")
        print(f"  {'─'*60}")
        print(f"    T_amb = {res['T_amb']:.3f} °C   T0 = {res['T0']:.3f} °C")
        print(f"    k     = {res['k']:.6f} ± {res['k_err']:.6f} s⁻¹")
        print(f"    R²    = {res['R2']:.6f}  |  RMSE = {res['RMSE']:.4f} °C  |  χ²ᵣ = {res['chi2_r']:.4f}")
        print(f"\n    Cp = {res['Cp']:.4f} ± {res['Cp_err']:.4f}  J/(g·K)  (IC {CONFIANZA*100:.0f}%)")
        if Cp_ref_val:
            print(f"    Cp tablas      = {Cp_ref_val:.4f} J/(g·K)")
            print(f"    Discrepancia   = {discrepancia:.2f}%   ({sigma_n:.2f} σ)")
 
    print(f"\n{'='*65}")
    print("  RESUMEN")
    print(f"{'='*65}")
    df_res = pd.DataFrame(resultados)
    df_res["Cp_ref"] = df_res["nombre"].map(CP_REFERENCIA)
    df_res["disc_%"] = ((df_res["Cp"] - df_res["Cp_ref"]).abs() / df_res["Cp_ref"] * 100).round(2)
    print(df_res[["nombre","k","k_err","R2","Cp","Cp_err","Cp_ref","disc_%"]].to_string(index=False))
 
    df_res.to_csv("resultados_calorimetro.csv", index=False)
    print("\n  Guardado en: resultados_calorimetro.csv")
```

```python
# ── Plot functions ────────────────────────────────────────────────────────────
# Requiere haber corrido calorimetro_analisis.py en el mismo kernel.
# Variables esperadas en scope: MATERIALES, CP_REFERENCIA, CONFIANZA,
#   hA, hA_err, C_cal, modelo_newton(), ajustar_exponencial(), cargar_csv()
# ─────────────────────────────────────────────────────────────────────────────

import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from scipy.stats import norm as scipy_norm

COLORES = ["#4C9BE8", "#E8A84C", "#E86B4C", "#7EE8A2", "#C97EE8"]

def _color(i):
    return COLORES[i % len(COLORES)]

def _calcular_Cp(nombre, cfg):
    tiempo_s, temp = cargar_csv(cfg["archivo"])
    popt, _, ci, _ = ajustar_exponencial(tiempo_s, temp)
    k, k_err       = popt[2], ci[2]
    Cp             = (hA / k - C_cal) / cfg["masa_g"]
    Cp_e           = np.sqrt((hA_err / (cfg["masa_g"] * k))**2 +
                             (-hA / (cfg["masa_g"] * k**2) * k_err)**2)
    return popt, k, k_err, Cp, Cp_e


def plot_curvas_enfriamiento():
    """T(t) datos + ajuste + residuos debajo, un subplot por material."""
    n    = len(MATERIALES)
    fig, axes = plt.subplots(2, n, figsize=(5 * n, 7),
                             gridspec_kw={"height_ratios": [3, 1]},
                             sharex="col")

    for i, (nombre, cfg) in enumerate(MATERIALES.items()):
        tiempo_s, temp = cargar_csv(cfg["archivo"])
        popt, _, ci, _ = ajustar_exponencial(tiempo_s, temp)
        T_fit          = modelo_newton(tiempo_s, *popt)
        color          = _color(i)

        ax_top = axes[0, i]
        ax_res = axes[1, i]

        ax_top.scatter(tiempo_s, temp, s=6, alpha=0.5, color=color, label="datos")
        ax_top.plot(tiempo_s, T_fit, lw=2, color="black", alpha=0.85,
                    label=f"k = {popt[2]:.4f} s⁻¹")
        ax_top.set_title(nombre, fontsize=13)
        ax_top.set_ylabel("T (°C)")
        ax_top.legend(fontsize=8)
        ax_top.grid(True)

        residuos = temp - T_fit
        ax_res.axhline(0, color="black", lw=0.8, ls="--")
        ax_res.scatter(tiempo_s, residuos, s=5, alpha=0.6, color=color)
        ax_res.set_ylabel("Res. (°C)", fontsize=8)
        ax_res.set_xlabel("Tiempo (s)")
        ax_res.grid(True)

    fig.suptitle("Curvas de enfriamiento — ajuste Newton", fontsize=15)
    plt.tight_layout()
    return fig


def plot_cp_comparacion():
    """Barras Cp experimental (± IC) vs. referencia CRC."""
    nombres, Cp_exp, Cp_err_list, Cp_ref_list = [], [], [], []

    for nombre, cfg in MATERIALES.items():
        _, _, _, Cp, Cp_e = _calcular_Cp(nombre, cfg)
        nombres.append(nombre)
        Cp_exp.append(Cp)
        Cp_err_list.append(Cp_e)
        Cp_ref_list.append(CP_REFERENCIA.get(nombre, np.nan))

    x, w   = np.arange(len(nombres)), 0.35
    colors = [_color(i) for i in range(len(nombres))]
    fig, ax = plt.subplots(figsize=(8, 5))

    ax.bar(x - w/2, Cp_exp, w, yerr=Cp_err_list, capsize=6,
           color=colors, alpha=0.85, label=f"Experimental (IC {int(CONFIANZA*100)}%)",
           error_kw={"lw": 2, "capthick": 2})
    ax.bar(x + w/2, Cp_ref_list, w,
           color="gray", alpha=0.4, edgecolor="black", lw=1.2,
           label="Referencia (CRC)")

    ax.set_xticks(x)
    ax.set_xticklabels(nombres, fontsize=12)
    ax.set_ylabel("$C_p$  (J g⁻¹ K⁻¹)", fontsize=12)
    ax.set_title("$C_p$ experimental vs. referencia", fontsize=14)
    ax.legend()
    ax.grid(True, axis="y")
    plt.tight_layout()
    return fig


def plot_residuos_hist():
    """Histograma de residuos normalizados + curva N(0,1) por material."""
    n    = len(MATERIALES)
    fig, axes = plt.subplots(1, n, figsize=(5 * n, 4))
    if n == 1:
        axes = [axes]

    for i, (nombre, cfg) in enumerate(MATERIALES.items()):
        tiempo_s, temp = cargar_csv(cfg["archivo"])
        popt, _, _, _  = ajustar_exponencial(tiempo_s, temp)
        res            = temp - modelo_newton(tiempo_s, *popt)
        res_norm       = (res - res.mean()) / res.std()

        axes[i].hist(res_norm, bins=15, density=True,
                     color=_color(i), alpha=0.7, edgecolor="black", lw=0.5)
        x_fit = np.linspace(res_norm.min(), res_norm.max(), 200)
        axes[i].plot(x_fit, scipy_norm.pdf(x_fit), color="black",
                     lw=1.8, ls="--", label="N(0,1)")
        axes[i].set_title(nombre, fontsize=11)
        axes[i].set_xlabel("σ")
        axes[i].set_ylabel("Densidad")
        axes[i].legend(fontsize=8)
        axes[i].grid(True)

    fig.suptitle("Distribución de residuos normalizados", fontsize=14)
    plt.tight_layout()
    return fig


def plot_discrepancia():
    """Discrepancia en % y en número de sigmas respecto a tablas CRC."""
    nombres, disc_pct, disc_sigma = [], [], []

    for nombre, cfg in MATERIALES.items():
        _, _, _, Cp, Cp_e = _calcular_Cp(nombre, cfg)
        ref = CP_REFERENCIA.get(nombre, np.nan)
        nombres.append(nombre)
        disc_pct.append(abs(Cp - ref) / ref * 100 if not np.isnan(ref) else np.nan)
        disc_sigma.append(abs(Cp - ref) / Cp_e if (not np.isnan(ref) and Cp_e > 0) else np.nan)

    colors = [_color(i) for i in range(len(nombres))]
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

    ax1.bar(nombres, disc_pct, color=colors, alpha=0.85, edgecolor="black", lw=0.8)
    ax1.axhline(5,  color="goldenrod", ls="--", lw=1.4, label="5 %")
    ax1.axhline(10, color="tomato",    ls="--", lw=1.4, label="10 %")
    ax1.set_ylabel("Discrepancia (%)")
    ax1.set_title("Error relativo vs. CRC")
    ax1.legend(fontsize=9)
    ax1.grid(True, axis="y")

    ax2.bar(nombres, disc_sigma, color=colors, alpha=0.85, edgecolor="black", lw=0.8)
    ax2.axhline(2, color="goldenrod", ls="--", lw=1.4, label="2 σ")
    ax2.axhline(5, color="tomato",    ls="--", lw=1.4, label="5 σ")
    ax2.set_ylabel("Discrepancia (σ)")
    ax2.set_title("Significancia estadística")
    ax2.legend(fontsize=9)
    ax2.grid(True, axis="y")

    fig.suptitle("Discrepancia experimental vs. referencia", fontsize=14)
    plt.tight_layout()
    return fig


# ── Llamadas ──────────────────────────────────────────────────────────────────

fig1 = plot_curvas_enfriamiento()
fig1.savefig("01_curvas_enfriamiento.png", dpi=300, bbox_inches="tight")

fig2 = plot_cp_comparacion()
fig2.savefig("02_cp_comparacion.png", dpi=300, bbox_inches="tight")

fig3 = plot_residuos_hist()
fig3.savefig("03_residuos_hist.png", dpi=300, bbox_inches="tight")

fig4 = plot_discrepancia()
fig4.savefig("04_discrepancia.png", dpi=300, bbox_inches="tight")

# Opcional: si ya no quieres que se abran las ventanas interactivas
# al correr el script (para que sea más rápido), puedes comentar o borrar plt.show()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# =====================================================================
# Parámetros experimentales base (Extraídos del informe)
# =====================================================================
m_c = 56.37        # Masa equivalente del calorímetro (g)
m_agua = 100.0     # Masa del agua inicialmente en el calorímetro (g)
M_eq = m_c + m_agua # Masa equivalente  b
total en agua (g)

T_0_agua = 15.97   # Temperatura inicial del sistema agua+calorímetro (°C)
T_0_muestra = 91.8 # Temperatura inicial de la muestra en ebullición (°C)

# Diccionario con los datos específicos de cada sólido
muestras = {
    'Aluminio': {'archivo': 'al.csv', 'masa': 15.7},
    'Laton': {'archivo': 'la.csv', 'masa': 56.8},
    'Hierro': {'archivo': 'fe.csv', 'masa': 22.1}
}

resultados = {}

# =====================================================================
# Procesamiento, Cálculo y Graficación
# =====================================================================
for nombre, datos in muestras.items():
    if not os.path.exists(datos['archivo']):
        print(f"Advertencia: No se encontró el archivo {datos['archivo']}.")
        continue
        
    df = pd.read_csv(datos['archivo'])
    # Convertir tiempo de ms a segundos para la gráfica
    df['tiempo_s'] = df['tiempo_ms'] / 1000.0
    
    # Metodología A: Te como el promedio de los últimos 15 registros (estado estacionario)
    Te_promedio = df['T_celsius'].tail(15).mean()
    
    # Metodología B: Te como el valor mínimo alcanzado en la serie (asíntota experimental)
    Te_minimo = df['T_celsius'].min()
    
    # Cálculo de la capacidad calorífica (Eq. 6 del informe)
    # c_x = [M_eq * (Te - T_01)] / [m_2 * (T_02 - Te)]
    c_x_A = (M_eq * (Te_promedio - T_0_agua)) / (datos['masa'] * (T_0_muestra - Te_promedio))
    c_x_B = (M_eq * (Te_minimo - T_0_agua)) / (datos['masa'] * (T_0_muestra - Te_minimo))
    
    resultados[nombre] = {
        'Te_estacionario': Te_promedio,
        'cx_estacionario': c_x_A,
        'Te_minimo': Te_minimo,
        'cx_minimo': c_x_B
    }
    
    # Generación de la gráfica individual
    plt.figure(figsize=(8, 5))
    plt.plot(df['tiempo_s'], df['T_celsius'], label=f'Datos NTC100K - {nombre}', color='#2c3e50', linewidth=2)
    plt.axhline(y=Te_promedio, color='#e74c3c', linestyle='--', label=f'$T_e$ (Promedio) = {Te_promedio:.2f} °C')
    
    plt.title(f'Evolución Térmica hacia el Equilibrio - {nombre}', fontsize=14)
    plt.xlabel('Tiempo (s)', fontsize=12)
    plt.ylabel('Temperatura (°C)', fontsize=12)
    plt.legend(loc='upper right')
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.tight_layout()
    
    # Guardar la figura de forma independiente
    nombre_figura = f'equilibrio_{nombre.lower()}.png'
    plt.savefig(nombre_figura, dpi=300)
    plt.close() # Cierra la figura para no saturar la memoria del notebook

# =====================================================================
# Impresión de Resultados
# =====================================================================
print("-" * 60)
print(f"{'Material':<12} | {'Te (°C)':<10} | {'c_x (cal/g°C) [Promedio]':<25}")
print("-" * 60)
for nombre, res in resultados.items():
    print(f"{nombre:<12} | {res['Te_estacionario']:<10.2f} | {res['cx_estacionario']:.4f}")
print("-" * 60)
```
