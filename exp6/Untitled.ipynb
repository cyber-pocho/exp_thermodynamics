```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import pandas as pd
```

```python
plt.style.use(["science","ieee"])
```

```python
df=pd.read_csv("touch.csv")
df.columns=[c.strip for c in df.columns]
V=df.iloc[:,0]
I=df.iloc[:,1]
T_high=df.iloc[:,2]
T_low=df.iloc[:,3]
```

```python
plt.figure(figsize=(10,4))
plt.scatter(I, V, s=5, label="Caracteristica I vs V para celda peltier", )
plt.xlabel("V (Voltaje) [$V$]")
plt.ylabel("I (Corriente) [$A$]")
plt.legend()
plt.show()
```

```python
delta_T=T_high-T_low
potencia=V*I
plt.figure(figsize=(10,5))
plt.scatter(potencia, delta_T, s=3, label=r"Potencia vs $ \Delta T$")
plt.scatter(delta_T, V, s=3, label=r"Potencia vs Voltaje")
plt.xlabel(r"Potencia (Watts) [$W$]")
plt.ylabel(r"Diferencia de temperatura ($\delta T$) [$^\circ C$]")
plt.legend()
plt.show()
```

```python
delta_T=T_high-T_low
potencia=V*I
plt.figure(figsize=(10,5))
plt.scatter(I, delta_T, s=3, label=r"$\Delta T$ vs Corriente")
plt.xlabel(r"Corriente")
plt.ylabel(r"Diferencia de temperatura ($\delta T$) [$^\circ C$]")
plt.legend()
plt.grid()
plt.show()
```
